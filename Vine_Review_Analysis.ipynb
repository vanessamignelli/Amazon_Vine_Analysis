```python
# import dependencies
import pandas as pd
```

```python
# read in vine table csv as df 
vine_table_df = pd.read_csv('vine_table.csv', low_memory=False)
vine_table_df.head()
```

```python
# create a new df to retrieve all rows where 'total_votes' are >=20
vine_tot_votes_df = vine_table_df.loc[vine_table_df['total_votes'] >= 20].reset_index(drop=True)
vine_tot_votes_df.head(10)
```

```python
# create a new df to retrieve all rows where # of 'helpful_votes'/'total_votes' are >= 50%
vine_helpful_votes_df = vine_tot_votes_df.copy()
vine_helpful_votes_df['helpful_votes_percentage'] = vine_helpful_votes_df['helpful_votes']/vine_helpful_votes_df['total_votes']
vine_helpful_votes_df = vine_helpful_votes_df.loc[vine_helpful_votes_df['helpful_votes_percentage'] >= 0.5].reset_index(drop=True)
vine_helpful_votes_df.head(10)
```

```python
# create a new df to retrieve all rows where 'vine' == Y
paid_reviews_df = vine_helpful_votes_df.loc[vine_helpful_votes_df['vine'] == 'Y'].reset_index(drop=True)
paid_reviews_df.head(10)
```

```python
# create a new df to retrieve all rows where 'vine' == N
unpaid_reviews_df = vine_helpful_votes_df.loc[vine_helpful_votes_df['vine'] == 'N'].reset_index(drop=True)
unpaid_reviews_df.head(10)
```

```python
# get total number of paid reviews
tot_reviews_paid = paid_reviews_df['review_id'].count()
tot_reviews_paid
```

```python
# get total number of 5 star paid reviews
five_star_paid = paid_reviews_df[(paid_reviews_df['star_rating'] >= 5)]
five_star_paid = five_star_paid['review_id'].count()
five_star_paid
```

```python
# get % of 5 star paid reviews
five_star_percentage = (five_star_paid/tot_reviews_paid) * 100
five_star_percentage
```

```python
# get total number of unpaid reviews
tot_reviews_unpaid = unpaid_reviews_df['review_id'].count()
tot_reviews_unpaid
```

```python
# get total number of 5 star unpaid reviews
five_star_unpaid = unpaid_reviews_df[(unpaid_reviews_df['star_rating'] >= 5)]
five_star_unpaid = five_star_unpaid['review_id'].count()
five_star_unpaid
```

```python
# get % of 5 star unpaid reviews
five_star_percentage_unpaid = (five_star_unpaid/tot_reviews_unpaid) * 100
five_star_percentage_unpaid
```
